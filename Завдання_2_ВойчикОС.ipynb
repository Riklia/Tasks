1. Дано: одновимірний масив

```python
arr = np.random.rand(1000)
```

Завдання: замінити всі елементи цього масиву, які більші за середнє цього масива на це середнє значення.

```python
new_arr = np.where(arr > arr.mean(), arr.mean(), arr)
print(new_arr)
```

2. Для масиву з попередньої задачі імплементувати сигмоїдну функцію:

$$y = \frac{1} {1 + e^{-x}}$$

```python
new_arr = 1/(1+np.exp(-arr))
print(new_arr)
```

3. Вирішити завдання про таблицю співробітників використовуючи numpy (вивести на екран інформацію про співробітників, прийнятих на роботу після 2012 року та з ЗП понад 2000; надрукувати можна без форматування)

```python
workers_table = np.array([
    [1, 'B', 1000, 2013],
    [2, 'B', 1500, 2014],
    [3, 'B', 1800, 2012],
    [4, 'A', 2500, 2016],
    [5, 'A', 3500, 2017],
    [6, 'A', 4500, 2011],
], dtype=object)
print(workers_table[(workers_table[:, 3] > 2012) & (workers_table[:, 2] > 2000)])
```

4. У таблиці співробітників з попередньої задачі, додати ще один стовпець, де записати різницю між ЗП співробітника та середньою ЗП усіх співробітників.

```python
workers_table = np.concatenate([workers_table, 
                                (workers_table[:,2] - workers_table[:,2].mean()).reshape((len(workers_table), 1))], axis=1)
print(workers_table)
```

5. У таблиці співробітників знайти та вивести на екран мінімальний рік прийому на роботу в кожному департаменті.

```python
for department in set(workers_table[:,1]):
    print(f"{department} - {workers_table[workers_table[:, 1] == department][:, 3].min()}")
```

6. Дана таблиця A

```python
A = np.array([
    [2, 5, 8, 56],
    [4, 9, 1, 23],
    [6, 12, 84, 55]
])
```

Завдання: нормалізувати колонки (зробити так, щоб довжина векторів у колонках була 1)

```python
normalized_A = np.zeros((3, 4))
for i in range(len(A)):
    normalized_A[i] = A[i, :]/np.sqrt(np.sum(A[i, :]**2))

print(normalized_A)
```

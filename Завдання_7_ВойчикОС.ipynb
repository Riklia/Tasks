1. Знайти значення m і b для яких функція набуває мінімального значення:

$$f = \frac{1}{n}\sum_{i=1}^{n}{(y_{i} - (b + mx_{i}))^{2}}$$

&emsp;&emsp;де x,y - деякі задані вектори

```python
np.random.seed(42)
x = np.random.normal(0, 5, 100)
y = 50 + 2 * x + np.random.normal(0, 2, len(x))
```

```python
def f(b, m):
    return 1/len(x) * sum((y - (b + m * x))**2)

def df(b, m):
    df_b = -2/len(x) * sum(y - (b + m * x))
    df_m = -2/len(x) * sum(x * (y - (b + m * x)))
    return np.array([df_b, df_m])
```

```python
bm0 = np.ones(2)
alpha = 0.03
epsilon = 0.00001


print(bm0, f(bm0[0], bm0[1]))

while True:
    b = bm0[0]
    m = bm0[1]
    bm = bm0 - alpha * df(b, m)
    if np.sum(np.abs(bm - bm0)) <= epsilon: 
        break
    else:
        bm0 = bm 
    print(bm, f(bm[0], bm[1]))

print('=' * 100)
print('Значення bm, при якому f набуває мінімального значення: ', bm)
print('Мінімум f: ', f(bm[0], bm[1]))
```

2. Знайти значення m і b для яких функція набуває мінімального значення:

$$f = \frac{1}{n}\sum_{i=1}^{n}{[(y_{i} - (b + mx_{i}))^{2}]} + m^2 + b^2$$

&emsp;&emsp;де x,y - вектори з 1-го завдання

```python
def f(b, m):
    return 1/len(x) * sum((y - (b + m * x))**2) + m**2 + b**2

def df(b, m):
    h = 1e-10
    f_b = (f(b + h, m) - f(b, m)) / h
    f_m = (f(b, m + h) - f(b, m)) / h
    return np.array([f_b, f_m])
```

```python
bm0 = np.ones(2)
alpha = 0.02
epsilon = 0.00001

print(bm0, f(bm0[0], bm0[1]))

while True:
    b = bm0[0]
    m = bm0[1]
    bm = bm0 - alpha * df(b, m)
    if np.sum(np.abs(bm - bm0)) <= epsilon: 
        break
    else:
        bm0 = bm 
    print(bm, f(bm[0], bm[1]))

print('=' * 100)
print('Значення bm, при якому f набуває мінімального значення: ', bm)
print('Мінімум f: ', f(bm[0], bm[1]))
```

3. Знайти значення а, для якого функція набуває мінімального значення

$$f = -\frac{1}{n}\sum_{i=1}^{n}{[y_{i}log(h(x_{i})) + (1 - y_{i})log(1 - h(x_{i}))]}$$

$$h = \frac{1} {1 + exp^{-ax}}$$

```python
np.random.seed(10)
num_observations = 100
x1 = np.random.normal(1, 1, 50)
x2 = np.random.normal(3, 1, 50)

x = np.concatenate([x1, x2])
y = np.concatenate([np.zeros_like(x1), np.ones_like(x2)])
```

```python
# h = lambda a: 1 / (1 + np.exp(-a * x))

def f(a):
    h = 1 / (1 + np.exp(-a * x))
    return -1/len(x) * sum(y * np.log(h) + (1 - y) * np.log(1 - h))

def df(a):
    h = 1e-10
    return (f(a + h) - f(a)) / h
```

```python
a0 = 1
alpha = 0.1
epsilon = 0.00001

print(a0, f(a0))

while True:
    a = a0
    a = a0 - alpha * df(a)
    if np.abs(a - a0) <= epsilon: 
        break
    else:
        a0 = a 
    print(a, f(a))

print('=' * 100)
print('Значення bm, при якому f набуває мінімального значення: ', a)
print('Мінімум f: ', f(a))
```

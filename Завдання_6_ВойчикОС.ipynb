1. Створити графіки для віку пасажирів та ціни білетів на одній області та на двох областях (одній над іншою).

```python
import pandas as pd
import matplotlib.pyplot as plt

titanic_data = pd.read_csv('Заняття6/Files/TitanicDataset.csv', sep=',', header=0)

titanic_data.plot(x="PassengerId", y=["Age", "Fare"])
plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(titanic_data['PassengerId'], titanic_data['Age'])
ax1.set_ylim(0, 100)
ax1.set_title('Age')
ax1.grid(True)

ax2 = fig.add_subplot(2, 1, 2)
ax2.plot(titanic_data['PassengerId'], titanic_data['Fare'])
ax2.set_ylim(0, titanic_data['Fare'].max()+100)
ax2.set_title('Fare')
ax2.grid(True)

plt.tight_layout(h_pad = 2, w_pad = -0.2)
plt.show()
```

2. Створити стовпчикову діаграму для статі пасажирів (кількість пасажирів тої чи іншої статі) та кількості що вижили залежно від статі.

```python
passengers_sex_data = titanic_data.groupby(['Sex'])['PassengerId'].count()
survived_sex_data = titanic_data.groupby(by="Sex")['Survived'].sum()

fig, ax = plt.subplots(figsize=(5,5))

ax.bar(x = survived_sex_data.index, 
       height = survived_sex_data.values, width=0.2, color='blue', label='Survived')
ax.set_title('Survived')

ax.bar(x = passengers_sex_data.index, 
       height = passengers_sex_data.values, width = 0.1, 
       color = 'orange', align='edge', label='Passengers')
ax.legend()
ax.set_title('Passenger')
ax.set_xlabel('Sex')
ax.set_ylabel('Count')
plt.show()
```

3. Створити стовпчикову діаграму для класу квитка пасажирів (кількість пасажирів відповідного класу) та кількості що вижили залежно від статі.

```python
passengers_class_data = titanic_data.groupby(['Pclass'])['PassengerId'].count()
survived_class_data = titanic_data.groupby(["Pclass"])['Survived'].sum()

fig, ax = plt.subplots(figsize=(10,5))

ax.bar(x = survived_class_data.index, 
       height = survived_class_data.values, width=0.15, 
       color='blue', label='Survived')
ax.set_title('Survived')


ax.bar(x = passengers_class_data.index, 
       height = passengers_class_data.values, width = 0.095, 
       color = 'orange', align='edge', label='Passengers')
ax.legend()
ax.set_title('Passengers')
ax.set_xlabel('Class')
ax.set_ylabel('Count')
plt.show()
```

4. Створити кругову діаграму по кількості пасажирів за місцем посадки.

```python
passengers_class_data.plot(kind="pie", label="")
plt.title("Кількість пасажирів залежно від місця посадки")
plt.show()
```

5. Побудуйте графіки функцій на відповідних проміжках:  
$y=x^3-2x+1$, $-4≤x≤4$  
$y=cos(cos(3x))-sin(sin(x))$, $-4 \pi≤x≤4\pi$  
$y=x^2+{{1}\over{x}}$, $-4≤x≤4$


```python
import numpy as np

x1 = np.arange(-4, 4, 0.1)
f1 = x1**3 - 2*x1 + 1

x2 = np.arange(-4*np.pi, 4*np.pi, 0.1)
f2 = np.cos(np.cos(3*x2))-np.sin(np.sin(x2))

x3 = np.arange(-4, 4, 0.1)
f3 = x3**2 + 1/x3

plt.grid(True)
plt.xlabel(u'x')
plt.ylabel(u'f(x)')
plt.title(u'Графіки')
plt.plot(x1, f1, label = u'$x^3 - 2x + 1$')
plt.legend(loc='lower left')
plt.show()

plt.grid(True)
plt.xlabel(u'x')
plt.ylabel(u'f(x)')
plt.plot(x2, f2, label = u'$cos(cos(3x))-sin(sin(x))$')
plt.legend(loc='lower left')
plt.show()

plt.grid(True)
plt.xlabel(u'x')
plt.ylabel(u'f(x)')
plt.plot(x3, f3, label = u'$x^2 + 1/x$')
plt.legend(loc='lower left')
plt.show()

```
